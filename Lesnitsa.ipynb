```python
n=int(input())
a=list(map(int,input().split()))
def func(a,n):
    pr1=a[0]
    pr2=0
    for i in range (1,n):
        cur=max(pr1+a[i],pr2+a[i])
        pr2=pr1
        pr1=cur
    return pr1
print(func(a,n))
```
