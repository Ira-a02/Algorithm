```python
n, *A=map(int,input().split())
k, *B=map(int, input().split())
res=[]
def search (index, value):
    l, r = 0, n-1
    while (l<=r):
        m=l+(r-l)//2
        if index[m]==value:
            res.append(m+1)
            break
        elif index[m]>value: r=m-1
        else: l=m+1
    else:
        res.append(-1)

for i in B:
    search(A, i)

print(*res, sep=' ')
```
