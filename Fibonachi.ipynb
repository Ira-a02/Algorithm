```python
def fib(n):
    if n == 1 or n == 2:
        return 1
    prev = 1
    current = 1
    for i in range(3, n + 1):
        temp = current;
        current = prev + current;
        prev = temp;
    return current
def main():
    n = int(input())
    print(fib(n))


if __name__ == "__main__":
    main()
```

```python
def fib_digit(n):
    if n == 1 or n == 2:
        return 1
    prev = 1
    current = 1
    for i in range(3, n + 1):
        temp = current;
        current = (prev + current) % 10;
        prev = temp;
    return current

def main():
    n = int(input())
    print(fib_digit(n))


if __name__ == "__main__":
    main()
```

```python
def fib_mod(n, m):
    period = [0, 1]
    i = 2
    while i < m * 6:
        period.append((period[i - 1] + period[i - 2]) % m)
        if period[i] == 1 and period[i - 1] == 0:
            break
        i += 1
    return (period[n % (len(period) - 2)])


def main():
    n, m = map(int, input().split())
    print(fib_mod(n, m))


if __name__ == "__main__":
    main()
```
