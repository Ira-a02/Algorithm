```python
n=int(input())
a=list(map(int,input().split()))
def func(a):
    k=[0]*(len(a)+1)
    for i in range(1,len(a)+1):
        mmax=0
        for j in range(i):
            if (a[i-1]%a[j-1]==0) and(k[j]>mmax):
                mmax=k[j]
            k[i]=mmax+1
    return max(k)
print(func(a))
```
