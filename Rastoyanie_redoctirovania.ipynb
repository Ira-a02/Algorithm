```python
# put your python code here
a = input()
b = input()
m = len(a)
n = len(b)

def func(a,m,b,n):

    d=[[0 for i in range(n+1)] for j in range(m+1)]
    for i in range(1, m+1):
        d[i][0] = i
    for j in range(1, n+1):
        d[0][j] = j
    for i in range(1, m+1):
        for j in range(1, n+1):
            if (a[i-1] == b[j-1]):
                cost = 0
            else:
                cost = 1

            d[i][j]=min(d[i-1][j]+1,d[i][j-1]+1,d[i-1][j-1]+cost)
    return d[m][n]
print(func(a,m,b,n))



```
